```python
import pandas as pd
```

MovieLens

```python
movie_lens = "data/ml-latest-small"
links = pd.read_csv(f"{movie_lens}/links.csv")
movies = pd.read_csv(f"{movie_lens}/movies.csv")
ratings = pd.read_csv(f"{movie_lens}/ratings.csv")
tags = pd.read_csv(f"{movie_lens}/tags.csv")
```

```python
print(f"movies: {movies.shape}")
print(f"ratings: {ratings.shape}")
print(f"tags: {tags.shape}")
print(f"links: {links.shape}")
```

Movies

```python
movies.head()
```

```python
movies["year"] = movies["title"].str.extract(r"\((\d{4})\)")
```

```python
movies["year"] = movies["year"].astype("Int64")
```

```python

movies["title"] = movies["title"].str.replace(r"\s*\(\d{4}\)$", "", regex=True).str.strip()
```

```python
# movies['year_missing'] = movies['year'].isna().astype(int)
```

```python
print(movies.isnull().sum())
```

```python
movies.head()
```

Ratings

```python
print(ratings.isnull().sum())
```

```python
ratings["timestamp"] = pd.to_datetime(ratings["timestamp"], unit="s")
ratings["year"], ratings["month"]  = ratings["timestamp"].dt.year, ratings["timestamp"].dt.month
```

```python
ratings.duplicated().sum()
```

```python
ratings.head()
```

Links

```python
print(links.isnull().sum())
```

```python
links["tmdbId"] = links["tmdbId"].astype("Int64")
```

```python
links["imdbId"] = links["imdbId"].astype("Int64")
```

```python
links.head()
```

```python
links.duplicated().sum()
```

Tags

```python
print(tags.isnull().sum())
```

```python
tags["timestamp"] = pd.to_datetime(tags["timestamp"], unit="s")
tags["year"], tags["month"]  = tags["timestamp"].dt.year, tags["timestamp"].dt.month
```

```python
tags.head()
```

```python
movies.to_csv("data/ml-latest-small/movies_cleaned.csv", index=False)
ratings.to_csv("data/ml-latest-small/ratings_cleaned.csv", index=False)
links.to_csv("data/ml-latest-small/links_cleaned.csv", index=False)
tags.to_csv("data/ml-latest-small/tags_cleaned.csv", index=False)
```
