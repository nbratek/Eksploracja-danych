Credits

```python
credits_md.head()
```

```python
import ast

credits_md['cast'] = credits_md['cast'].apply(ast.literal_eval)
credits_md['crew'] = credits_md['crew'].apply(ast.literal_eval)

def get_cast(cast_list, n=5):
    return [actor['name'] for actor in cast_list[:n]]

def get_director(crew_list):
    for person in crew_list:
        if person['job'] == 'Director':
            return person['name']
    return None

def get_writers(crew_list):
    writers = []
    for person in crew_list:
        if person['department'] == 'Writing':
            writers.append(person['name'])
    return writers[:3]

credits_md['cast_main'] = credits_md['cast'].apply(get_cast)
credits_md['director'] = credits_md['crew'].apply(get_director)
credits_md['writers'] = credits_md['crew'].apply(get_writers)

credits_md['cast_main'] = credits_md['cast_main'].apply(lambda x: " ".join(x))
credits_md['writers'] = credits_md['writers'].apply(lambda x: " ".join(x))

credits_md = credits_md[['id', 'cast_main', 'director', 'writers']]
```

```python
print(credits_md.isnull().sum())
```

```python
credits_md["director"] = credits_md["director"].fillna("unknown")
```

```python
credits_md.duplicated().sum()
```

```python
credits_md = credits_md.drop_duplicates()
credits_md.duplicated().sum()
```

Keywords

```python
keywords_md.head()
```

```python
import ast

keywords_md["keywords"] = keywords_md["keywords"].apply(ast.literal_eval)

def get_keywords(keyword_list):
    words = []
    for item in keyword_list:
        word = item["name"].lower().replace(" ", "_")
        words.append(word)
    return words

keywords_md["keywords_clean"] = keywords_md["keywords"].apply(get_keywords)
keywords_md["keywords_clean"] = keywords_md["keywords_clean"].apply(lambda x: " ".join(x))

keywords_md = keywords_md[["id", "keywords_clean"]]
```

Movies metadata

```python
movies_metadata_md.head()
```

```python
movies_metadata_md = movies_metadata_md[movies_metadata_md['id'].str.isnumeric()]
movies_metadata_md['id'] = movies_metadata_md['id'].astype("Int64")
```

```python
import ast

def parse_json(x):
    try:
        return ast.literal_eval(x)
    except:
        return []

def get_names(list_obj):
    names = []
    for item in list_obj:
        if isinstance(item, dict) and "name" in item:
            names.append(item["name"].lower().replace(" ", "_"))
    return " ".join(names)

def get_collection(x):
    try:
        obj = ast.literal_eval(x)
        return obj["name"].lower().replace(" ", "_")
    except:
        return ""

movies_metadata_md["genres"] = (
    movies_metadata_md["genres"]
    .apply(parse_json)
    .apply(get_names)
)

movies_metadata_md["production_companies"] = (
    movies_metadata_md["production_companies"]
    .apply(parse_json)
    .apply(get_names)
)

movies_metadata_md["production_countries"] = (
    movies_metadata_md["production_countries"]
    .apply(parse_json)
    .apply(get_names)
)

movies_metadata_md["spoken_languages"] = (
    movies_metadata_md["spoken_languages"]
    .apply(parse_json)
    .apply(get_names)
)

movies_metadata_md["belongs_to_collection"] = (
    movies_metadata_md["belongs_to_collection"]
    .apply(get_collection)
)
```

```python
movies_metadata_md = movies_metadata_md.drop_duplicates()
movies_metadata_md.duplicated().sum()
```

```python
print(movies_metadata_md.isnull().sum())
```

```python
text_cols = ["homepage", "overview", "tagline", "status", "original_language", "poster_path"]
movies_metadata_md[text_cols] = movies_metadata_md[text_cols].fillna("unknown")
```

```python
int_cols = ["budget", "revenue", "runtime", "vote_count"]

for col in int_cols:
    movies_metadata_md[col] = movies_metadata_md[col].astype("Int64")

float_cols = ["vote_average", "popularity"]
for col in float_cols:
    movies_metadata_md[col] = movies_metadata_md[col].astype("float64")
```
